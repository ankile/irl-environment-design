```python
import matplotlib.pyplot as plt
import numpy as np
```

Low load: few behaviors, regular decision boundary

```python
dx= 0.01
x = np.arange(0,2*np.pi, dx) 
y = np.arange(0,2*np.pi, dx) 


Z = np.zeros(shape=(len(x), len(x)))

for idx_x, _x in enumerate(x):
    for idx_y, _y in enumerate(y):
        Z[idx_x, idx_y] = _x+_y

# Z = Z-np.min(Z)
Z = (Z - np.min(Z)) / (np.max(Z) - np.min(Z))
                        
Z = np.round(Z)
plt.matshow(Z, cmap ="tab20c",)
plt.xticks([], [])
plt.xlabel("$\gamma$")
plt.yticks([], [])
plt.ylabel("$p$")
```

Medium load: few behaviors but complex decision boundary

```python
dx= 0.01
x = np.arange(0,2*np.pi, dx) 
y = np.arange(0,2*np.pi, dx) 


Z = np.zeros(shape=(len(x), len(x)))

for idx_x, _x in enumerate(x):
    for idx_y, _y in enumerate(y):
        Z[idx_x, idx_y] = np.sin(2*_y) + np.sin(3*_x)

# Z = Z-np.min(Z)
(Z - np.min(Z)) / (np.max(Z) - np.min(Z))
                        
Z = np.round(Z)
plt.matshow(Z, cmap ="tab20c",)
plt.xticks([], [])
plt.xlabel("$\gamma$")
plt.yticks([], [])
plt.ylabel("$p$")
```

High load: many behaviors and complex decision boundaries

```python
dx= 0.01
x = np.arange(0,2*np.pi, dx) 
y = np.arange(0,2*np.pi, dx) 


Z = np.zeros(shape=(len(x), len(x)))

for idx_x, _x in enumerate(x):
    for idx_y, _y in enumerate(y):
        Z[idx_x, idx_y] = np.sin(_x) + np.cos(_y) + np.cos(0.5*_x) + np.sin(0.5*_y) + np.cos(2*_x) + np.sin(2*_y)

Z = np.round(Z)
plt.matshow(Z, cmap ="tab20c",)
plt.xticks([], [])
plt.xlabel("$\gamma$")
plt.yticks([], [])
plt.ylabel("$p$")
```
